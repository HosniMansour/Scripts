```python
import pandas as pd
```

```python
# Let's Respect Privacy :v !
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
colnames=['CIN', 'Name', 'First', 'FirstC'] 
df1 = pd.read_csv('1.csv', names=colnames, header=0)
scaler.fit(df1[['CIN']])
df1[['CIN']] = scaler.transform(df1[['CIN']])
df1.head()
```

```python
colnames=['CIN', 'Name', 'Second', 'SecondC'] 
df2 = pd.read_csv('2.csv', names=colnames, header=0)
df2[['CIN']] = scaler.transform(df2[['CIN']])
df2.head()
```

```python
df = pd.merge(df1, df2,on='CIN', how='outer')
df.head()
```

```python
df['FirstC'].fillna(0, inplace=True)
df['First'].fillna((df['First'].mean()), inplace=True)
```

```python
df['Moy'] = ( df.First + df.Second) /2
df['MoyC'] = ( df.FirstC + df.SecondC) /2
df.head()
```

```python
colnames=['CIN','Name_y', 'First', 'Second', 'FirstC', 'SecondC', 'Moy' ,'MoyC'] 
```
